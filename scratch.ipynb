```python
import os
import googleapiclient.discovery
from googleapiclient.discovery import build
from dotenv import load_dotenv
```

```python
# Load YOUTUBE_API_KEY from .env file
load_dotenv()  
API_KEY = os.getenv('YOUTUBE_API_KEY')
```

```python
def get_comment_threads(video_id, api_key):
    youtube = googleapiclient.discovery.build(
        "youtube", "v3", developerKey=api_key)
    
    request = youtube.commentThreads().list(
        part="snippet",
        videoId=video_id,
        maxResults=100
    )
    response = request.execute()
    
    return response
```

```python
video_id = "BOZ99Zw8dmw"

comment_threads = get_comment_threads(video_id, API_KEY)

for item in comment_threads["items"]:
    username = item["snippet"]["topLevelComment"]["snippet"]["authorDisplayName"]
    print(username)
```

```python
def get_all_comment_threads(video_id, api_key):
    youtube = googleapiclient.discovery.build(
        "youtube", "v3", developerKey=api_key
    )
    
    all_comments = []
    next_page_token = None
    
    while True:
        request = youtube.commentThreads().list(
            part="snippet",
            videoId=video_id,
            maxResults=100,
            pageToken=next_page_token
        )
        response = request.execute()
        
        all_comments.extend(response.get("items", []))
        
        next_page_token = response.get("nextPageToken")
        
        if not next_page_token:
            break
    
    return all_comments
```

```python
video_id = "BOZ99Zw8dmw"

comment_threads = get_all_comment_threads(video_id, API_KEY)

print(len(comment_threads))
```

```python
for item in comment_threads:
    username = item["snippet"]["topLevelComment"]["snippet"]["authorDisplayName"]
    print(username)
```

```python
def get_all_comment_threads_and_replies(video_id, api_key):
    youtube = build("youtube", "v3", developerKey=api_key)
    
    all_comments = []
    next_page_token = None
    
    while True:
        request = youtube.commentThreads().list(
            part="snippet,replies",
            videoId=video_id,
            maxResults=100,
            pageToken=next_page_token
        )
        response = request.execute()
        
        # Add top-level comments
        all_comments.extend(response.get("items", []))
        
        # Add replies to the comments
        for item in response.get("items", []):
            if 'replies' in item:
                all_comments.extend(item['replies']['comments'])
        
        next_page_token = response.get("nextPageToken")
        
        if not next_page_token:
            break
    
    return all_comments
```

```python
video_id = "BOZ99Zw8dmw"

comment_threads = get_all_comment_threads_and_replies(video_id, API_KEY)

print(len(comment_threads))
```
