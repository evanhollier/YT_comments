```python
import pandas as pd
from io import StringIO
import matplotlib.pyplot as plt
```

```python
import numpy as np
print(np.__version__)
```

```python
# RLM
# 52 Occurences of non UTF-8 character

with open("Channels/RedLetterMedia.csv", 'r', encoding='utf-8', errors='replace') as file:
    df = pd.read_csv(StringIO(file.read()), header=None)

    raw_length = len(df)
    print(raw_length)
   
    unique_length = len(set(df[0].str.strip()))
    print(unique_length)
```

```python
stages = ['Raw data', 'Handle non UTF-8', 'Remove duplicates']
sizes = [raw_length+52, raw_length, unique_length]
#print(sizes)

removed = [0, sizes[0] - sizes[1], sizes[1] - sizes[2]]
#print(removed)

fig, ax = plt.subplots()

for i in range(len(sizes)):
    ax.bar(stages[i], sizes[i], color='black')
    
    ax.text(stages[i], sizes[i] / 2, f'{sizes[i]}', ha='center', va='center', color='white', fontsize=12)

    if removed[i] > 0:
        ax.bar(stages[i], removed[i], bottom=sizes[i], color='red')
        ax.text(stages[i], sizes[i] + removed[i] - 0.02 * removed[i], f'-{removed[i]}', ha='center', va='top', color='white', fontsize=12)

ax.set_ylabel('Number of Rows')
ax.set_title('RedLetterMedia Data Size')

plt.show()
```
